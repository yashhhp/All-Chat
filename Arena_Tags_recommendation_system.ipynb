```python
df["SerialNo"]=df.index
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv= CountVectorizer()
count_matrix= cv.fit_transform(df['tags'])
```

```python
print(count_matrix)
```

```python
cos_sim= cosine_similarity(count_matrix)
print(cos_sim)
```

```python
def get_name_from_serialno(serial):
    return df[df.SerialNo==serial]['name'].values[0]

def get_serialno_from_name(name):
    return df[df.name==name]['SerialNo'].values[0]
```

```python
prob_user_likes= "Strategic Hopscotch"
prob_serialno= get_serialno_from_name(prob_user_likes)
#print(prob_serialno)
similar_probs= list(enumerate(cos_sim[prob_serialno]))
```

```python
sorted_similar_probs= sorted(similar_probs,key= lambda x:x[1], reverse=True)[1:]
```

```python
i=0
for element in sorted_similar_probs:
    print(get_name_from_serialno(element[0]))
    i=i+1
    if i>10:
        break
```
